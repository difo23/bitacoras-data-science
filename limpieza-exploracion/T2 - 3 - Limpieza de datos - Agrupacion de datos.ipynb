# Agregacion de datos por categoria

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
genero = ['Male','Femele']
income = ['Poor', 'Middle Class', 'Rich']
```

```python
n = 500
gender_data = []
income_data = []
```

```python
for i in range(0, n):
    gender_data.append(np.random.choice(genero))
    income_data.append(np.random.choice(income))
```

```python
height = 160 +15 *np.random.randn(n)
weight = 125 +25 *np.random.randn(n)
age = 26 + 9 *np.random.randn(n)
incomes = 22000 + 10000 *np.random.randn(n)
```

```python
data = pd.DataFrame({
    'height': height,
    'weight': weight,
    'age': age.astype(int),#convertir todo un narray en enteros
    'incomes': incomes,
    'gender': gender_data,
    "Level Economic": income_data
    
})
```

```python
data.head()
```

## Agrupacion de datos:

```python
grouped_gender =  data.groupby('gender')
```

```python
grouped_gender
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group('Femele')
```

```python
doble_group = data.groupby(['gender',"Level Economic"])
```

```python
doble_group.aggregate({
    "incomes": np.sum,
    "age":np.mean,
    "height":np.std
})
```

```python
doble_group.aggregate({
    
    "age":np.mean,
    "height": lambda hght: (np.mean(hght))/np.std(hght) 
})
```

```python
doble_group.aggregate([np.sum, np.mean, np.std])
```

```python
doble_group.aggregate([lambda x: np.mean(x)/np.std(x)])
```

### Transformacion 

```python
zscore = lambda x: (x-x.mean())/x.std()
```

```python
zgroup = doble_group.transform(zscore)
```

```python
zgroup
```

```python
plt.hist(zgroup['age'])
```
