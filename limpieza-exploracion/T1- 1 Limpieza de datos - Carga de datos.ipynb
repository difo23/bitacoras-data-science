######  Recorrer el documento o file obtenido por open() linea por linea y crear un diccionarios de arreglos:

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
for line in data3:
    values = line.strip().split(',')
    for item in range(len(cols)):
        main_dict[cols[item]].append(values[item])
    counter += 1
    
print("El dataset tiene columnas %d y %d filas"%(ncols, counter))
```

###### Usar urllib3


```python
#Crear funcion para descarga de datos usando urllib3

def downloadURLGet (url, filename, destinationpath = '../python-ml-course/datasets' , delimitado= '\n', sep= ',', decode= 'utf-8', extencion= 'csv'):
    http = urltools.PoolManager()
    response = http.request('GET', url)
    response.status
    res_data = response.data
    
    data_str = res_data.decode(decode)
    data_str = data_str.strip().split(delimitado)
    cols = data_str[0]

    cols = cols.split(sep)
    diccionario= {}
    for col in cols:
        diccionario[col]= []
    for line in data_str[1:]:
        values = line.strip().split(sep)
        for item in range(len(cols)):
            diccionario[cols[item]].append(values[item])

                                      
    df_url = pd.DataFrame(diccionario)
    file = os.path.join(destinationpath, filename+"."+extencion)
    
    if(extencion == 'xls'):
        df_url.to_excel(file)

    if(extencion == 'json'):
        df_url.to_json(file)
    
    if(extencion == 'xlsx'): 
        df_url.to_excel(file)
    
    if(extencion == 'csv'):
        df_url.to_csv(file, sep='\t', encoding='utf-8')
                              
    print('Se ha creado el fichero en :'+file)
    
    return df_url
```
